```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.constants import pi ,sigma,c,h,k, m_e, hbar, e
```

```python
#spherically-symmetric potential well of shape  φ(r)/c2 = 1 − exp(−1/2(r/rs)^2),
#Where I would not include rs under because of algebra
r = np.arange(-5, +5.1, 1, 'float')
def SSP(r):
    return 1-(np.exp(-0.5*(r)**2))
```

```python
SSP(r)
```

```python
plt.plot(r, SSP(r)) 
plt.xlabel('Distance r')
plt.ylabel('Gravitational redshift')
plt.title('Potential well')
```

```python
deltaZ = SSP(+5)-SSP(r)
print(deltaZ)
```

```python
plt.plot(r, deltaZ)
plt.xlabel('Distance r')
plt.ylabel('Gravitational redshift')
plt.title('Gravitational redshift as viewed by an observer +5r away')
```
