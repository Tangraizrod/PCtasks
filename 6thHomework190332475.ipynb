```python
import matplotlib.pyplot as plt
import numpy as np
import cmath as cmath
from scipy.constants import pi ,sigma,c,h,k, m_e, hbar, e
```

```python
Temp = np.linspace(3200, 4600, 8)
print(Temp)
```

```python
#we know from part b an expression for nh+np = ntot, where we found out that ntot= (1.6/(2.725^3))*T^3,
#which we can put inside the
constant = np.exp(-((13.6*e)/(k*Temp)))
def Saha_equation(Temp):
    return (((2.725**3)/(1.6*Temp**3))*(((m_e*k*Temp)/(2*pi*hbar**2))**1.5))*constant
y = Saha_equation(Temp)
print(y)
```

```python
#hence here I would like to calculate the xe, but the logic is that our y (saha equation) would always be positive b and
# negative c in the quadratic formula!, hence we can then use the quadratic for each saha value to find xe, and also we would
# like to only have the positive term +, hence I wont include
xe = (-y+np.sqrt(y**2-4*(-y)))/2 #we put -y because when we rearrange it is actually a negative term!

print(xe)
plt.plot(Temp, xe)
plt.xlabel('Temperature in Kelvin')
plt.ylabel('Ionisation fraction')
plt.axhline(0.5, color="black", linestyle="dashed")
plt.title('Plot of xe against T')
```
